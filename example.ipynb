# Example for effective Pore Size Analysis


```python
import MembraneAnalysisToolbox.EffectivePoreSizeAnalysis as mat
```

This is an example usage of the Effective Pore Size Analysis (part of the Membrane Analysis Toolbox)

1. Create the class with all the parameters needed for the analysis

```python
Analysis3nm = mat.EffectivePoreSizeAnalysis( 
    topology_file = '/bigpool/data/projects/Carbon_pores_Sofia/correctdensity/hex/3nm_1musonly/simulation_1/topol.tpr', 
    trajectory_file = '/bigpool/data/projects/Carbon_pores_Sofia/correctdensity/hex/3nm_1musonly/simulation_1/traj.xtc',
    membrane_resnames = ['C'],
    solvent_resnames = ['HEX', 'DOD'],
    y_middle = 40, 
    y_range = 10, 
    z_min = 150, 
    z_max = 300,
    verbose = True
)
```

2. calculate the effective pore size using this analysis class

```python
print("calculated effective pore size (3nm) is: " + str(Analysis3nm.calculate_effective_pore_size() / 10))
```

3. analyse, find and change possible constraints

```python
Analysis3nm._findConstraints()
Analysis3nm.analyseConstraints()
```

4. plot the results for verification and visualisation

```python
Analysis3nm.plot()
```
